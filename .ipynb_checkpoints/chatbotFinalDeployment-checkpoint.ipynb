# Load the Dataset

```python
from datasets import load_dataset

# Load the dataset with streaming enabled
ds = load_dataset("bitext/Bitext-customer-support-llm-chatbot-training-dataset", streaming=True)

# Preview the dataset
# Iterate over the dataset and fetch the first 5 records
for i, item in enumerate(ds['train']):
    print(item)
    if i == 4:  # Stop after printing 5 items
        break
```

## Extract Questions and Labels

```python
#Extract 'question', 'label', and 'response' fields from the dataset
labeled_data = [{'question': item['instruction'], 'label': item['intent'], 'response': item['response']} for item in ds['train']]

# Preview the labeled data
print(labeled_data[:5])  # Print the first 5 records
```

# Group Questions

 ## By Label

```python
from collections import defaultdict

# Group questions by their labels
grouped_data = defaultdict(list)
for item in labeled_data:
    question = item['question']
    label = item['label']
    grouped_data[label].append(question)

# Print the grouped data to verify
for label, questions in grouped_data.items():
    print(f"Label: {label}")
    print(f"Sample Questions: {questions[:5]}")  # Show first 5 questions per label
    print("-" * 50)
```

## by TF-IDF and Cosine Similarity

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from collections import defaultdict

# Function to group similar questions within each label
def group_similar_questions(questions):
    # Convert questions to TF-IDF vectors
    vectorizer = TfidfVectorizer()
    question_vectors = vectorizer.fit_transform(questions)
    
    # Calculate cosine similarity between all pairs of questions
    similarity_matrix = cosine_similarity(question_vectors)
    
    # Initialize clusters for similar questions
    clusters = defaultdict(list)
    visited = set()
    
    for i, question in enumerate(questions):
        if i not in visited:
            cluster = [question]
            visited.add(i)
            for j, similarity in enumerate(similarity_matrix[i]):
                if i != j and similarity > 0.7:  # Threshold for similarity, e.g., 0.7
                    cluster.append(questions[j])
                    visited.add(j)
            clusters[f'Cluster_{i}'] = cluster
    
    return clusters

# Apply the grouping within each label
all_clusters = defaultdict(dict)
for label, questions in grouped_data.items():
    print(f"Processing Label: {label}")
    clusters = group_similar_questions(questions)
    all_clusters[label] = clusters

# Print the clusters for each label
for label, clusters in all_clusters.items():
    print(f"Label: {label}")
    for cluster_name, cluster_questions in clusters.items():
        print(f"  {cluster_name}:")
        for question in cluster_questions:
            print(f"    {question}")
    print("-" * 50)
```

# Data Preprocessing

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer

# Download NLTK stopwords and WordNetLemmatizer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize the lemmatizer and stop words
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Preprocessing function
def preprocess_text(text):
    # Lowercasing
    text = text.lower()
    
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    
    # Tokenization
    tokens = word_tokenize(text)
    
    # Remove stopwords
    tokens = [word for word in tokens if word not in stop_words]
    
    # Lemmatization
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    # Remove leading and trailing whitespace for each token (although tokenization typically handles this)
    tokens = [token.strip() for token in tokens]
    
    # Join tokens back into a single string
    preprocessed_text = ' '.join(tokens)
    
    return preprocessed_text

# Apply preprocessing to all questions
for label, questions in grouped_data.items():
    grouped_data[label] = [preprocess_text(question) for question in questions]

# Preview the preprocessed data
for label, questions in grouped_data.items():
    print(f"Label: {label}")
    print(f"Sample Preprocessed Questions: {questions[:5]}")
    print("-" * 50)
```

# Data Splitting

```python
from sklearn.model_selection import train_test_split

# Create a dataset for each label with preprocessed questions and their labels
all_data = []
all_labels = []
for label, questions in grouped_data.items():
    for question in questions:
        all_data.append(question)
        all_labels.append(label)

# Split the data into training and testing sets (80% training, 20% testing)
train_data, test_data, train_labels, test_labels = train_test_split(all_data, all_labels, test_size=0.2, random_state=42)

# Verify the split
print(f"Training data: {len(train_data)} samples")
print(f"Testing data: {len(test_data)} samples")
```

# Model Training

## Algorithm 1: K-Nearest Neighbors (KNN)

```python
# Contribute By: BEH KAI FENG
# Install required libraries 
%pip install scikit-learn

# Import necessary libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import pandas as pd

# Convert the textual data into numerical features using TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the training data into numerical features (TF-IDF)
train_features = tfidf_vectorizer.fit_transform(train_data)

# Transform the test data using the same vectorizer (do not refit)
test_features = tfidf_vectorizer.transform(test_data)

# Initialize the KNN model
knn_model = KNeighborsClassifier(n_neighbors=10)  # You can change the number of neighbors as needed

# Train the KNN model on the training data
knn_model.fit(train_features, train_labels)

# Predict labels for the training and test data
train_predictions_knn = knn_model.predict(train_features)
test_predictions_knn = knn_model.predict(test_features)

# Calculate accuracy for both training and test data
train_accuracy_knn = accuracy_score(train_labels, train_predictions_knn)
test_accuracy_knn = accuracy_score(test_labels, test_predictions_knn)

# Print accuracy results
print(f"KNN Training Accuracy: {train_accuracy_knn * 100:.2f}%")
print(f"KNN Testing Accuracy: {test_accuracy_knn * 100:.2f}%")

# Print classification report for training data
print("\nClassification Report (Train Data - KNN):")
print(classification_report(train_labels, train_predictions_knn))

# Print classification report for test data
print("\nClassification Report (Test Data - KNN):")
print(classification_report(test_labels, test_predictions_knn))

# Example of Metrics Comparison Table for KNN
metrics = {
    'Model': ['KNN'],
    'Accuracy': [test_accuracy_knn],
    'Precision': [precision_score(test_labels, test_predictions_knn, average='weighted')],
    'Recall': [recall_score(test_labels, test_predictions_knn, average='weighted')],
    'F1-Score': [f1_score(test_labels, test_predictions_knn, average='weighted')]
}
```

### a. Accuracy

```python
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import KFold, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
import numpy as np
import pandas as pd


pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.width', 1000)  # Increase the display width to fit the table on one line

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(all_data)  # Transform text data into TF-IDF features
y = np.array(all_labels)  # Convert labels to a numpy array

kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Function to get cross-validation results
def get_cv_results(model, X, y):
    scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')
    return scores

# Define models
models = {
    'KNN': KNeighborsClassifier(n_neighbors=10),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
}

# Initialize results dictionary
results = {'Model': [], 'Fold 1': [], 'Fold 2': [], 'Fold 3': [], 'Fold 4': [], 'Fold 5': [], 'Mean(μ)': [], 'S.D.(σ)': []}

# Perform cross-validation for each model and store the results
for model_name, model in models.items():
    scores = get_cv_results(model, X, y)
    results['Model'].append(model_name)
    results['Fold 1'].append(scores[0])
    results['Fold 2'].append(scores[1])
    results['Fold 3'].append(scores[2])
    results['Fold 4'].append(scores[3])
    results['Fold 5'].append(scores[4])
    results['Mean(μ)'].append(scores.mean())
    results['S.D.(σ)'].append(scores.std())

# Convert results into a DataFrame
results_df = pd.DataFrame(results)

# Print accuracy metrics
print("Model Performance: Mean Accuracy and Standard Deviation Across Folds")
# Print results
print(results_df)

# Plotting the results
fig, ax = plt.subplots(figsize=(10, 6))

# Bar plot for mean accuracy with standard deviation as error bars
ax.bar(results_df['Model'], results_df['Mean(μ)'], yerr=results_df['S.D.(σ)'], capsize=5, color='skyblue',edgecolor='black')

# Add title and labels
ax.set_title('Model Accuracy with Standard Deviation Across 5 Folds', fontsize=16)
ax.set_xlabel('Model', fontsize=14)
ax.set_ylabel('Mean Accuracy', fontsize=14)
# Adjust y-axis limits
ax.set_ylim(0.9, 1.0)

# Display gridlines
ax.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import time
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

# Define the models
models = {
    'KNN': KNeighborsClassifier(n_neighbors=20),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(100,), max_iter=500, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
}

# Initialize dictionaries to store training and prediction times
training_times = {}
prediction_times = {}

# Loop through each model to measure time
for model_name, model in models.items():
    # Measure training time
    start_time = time.time()
    model.fit(train_features, train_labels)
    end_time = time.time()
    training_times[model_name] = end_time - start_time
    
    # Measure prediction time
    start_time = time.time()
    model.predict(test_features)
    end_time = time.time()
    prediction_times[model_name] = end_time - start_time

# Print training and prediction times
print("Training Times:")
for model_name, training_time in training_times.items():
    print(f"{model_name}: {training_time:.4f} seconds")

print("\nPrediction Times:")
for model_name, prediction_time in prediction_times.items():
    print(f"{model_name}: {prediction_time:.4f} seconds")

import matplotlib.pyplot as plt

# Data for plotting
models = list(training_times.keys())
training_times_values = list(training_times.values())
prediction_times_values = list(prediction_times.values())

# Plot Training Times
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.bar(models, training_times_values, color='skyblue')
plt.xlabel('Model')
plt.ylabel('Training Time (seconds)')
plt.title('Training Times')
plt.xticks(rotation=45, ha='right')

# Plot Prediction Times
plt.subplot(1, 2, 2)
plt.bar(models, prediction_times_values, color='salmon')
plt.xlabel('Model')
plt.ylabel('Prediction Time (seconds)')
plt.title('Prediction Times')
plt.xticks(rotation=45, ha='right')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plots
plt.show()
```

```python
from flask import Flask, render_template, request, jsonify
import joblib
from sklearn.feature_extraction.text import TfidfVectorizer
import re
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import random
import threading

# Initialize the lemmatizer and stop words
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Load the trained models and TF-IDF vectorizer from the file
mlp_model = joblib.load('neural_network_model.pkl')
knn_model = joblib.load('knn_model.pkl')
rf_model = joblib.load('random_forest_model.pkl')
dt_model = joblib.load('decision_tree_model.pkl')
vectorizer = joblib.load('tfidf_vectorizer.pkl')

# Preprocess function
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'[^\w\s]', '', text)
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    tokens = [token.strip() for token in tokens]
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text

# Preprocess input function
def preprocess_input(user_input):
    if user_input is None:
        raise ValueError("User input is None")
    preprocessed_input = preprocess_text(user_input)
    if preprocessed_input is None:
        raise ValueError("Preprocessed input is None")
    user_input_vec = vectorizer.transform([preprocessed_input])
    return user_input_vec

# Function to generate response based on the model prediction
def generate_response(prediction):
    responses = [item['response'] for item in labeled_data if item['label'] == prediction]
    if responses:
        return random.choice(responses)
    else:
        return "I'm sorry, I don't have an answer for that."

# Function to predict intent and confidence
def predict_intent(user_input, model):
    processed_input = preprocess_input(user_input)
    
    if model == 'neural_network':
        probabilities = mlp_model.predict_proba(processed_input)[0]
        prediction = mlp_model.classes_[probabilities.argmax()]
    elif model == 'knn':
        probabilities = knn_model.predict_proba(processed_input)[0]
        prediction = knn_model.classes_[probabilities.argmax()]
    elif model == 'random_forest':
        probabilities = rf_model.predict_proba(processed_input)[0]
        prediction = rf_model.classes_[probabilities.argmax()]
    elif model == 'decision_tree':
        probabilities = dt_model.predict_proba(processed_input)[0]
        prediction = dt_model.classes_[probabilities.argmax()]
    else:
        return "Invalid model selected", 0

    confidence = probabilities.max()
    return prediction, confidence

app = Flask(__name__)

@app.route("/")
def index():
    return render_template('chatbot.html')

@app.route("/get", methods=["POST"])
def chat():
    msg = request.form["msg"]
    selected_model = request.form.get("model", "neural_network")  # Default to 'neural_network'

    prediction, confidence = predict_intent(msg, selected_model)
    
    if confidence < 0.6:
        response = "I'm not very confident about this. Can you please provide more details?"
    else:
        response = generate_response(prediction)
    
    return jsonify({"response": response})

# Start Flask app in a thread so the notebook remains usable
def run_flask():
    app.run(debug=False, use_reloader=False)  # Disable reloader to prevent conflicts

# Run Flask in a separate thread
flask_thread = threading.Thread(target=run_flask)
flask_thread.start()
```
